```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/dimKlientai.csv'  # Replace with your file path
data = pd.read_csv(file_path, sep=';')

```

```python
# Converting income range to average income for simplicity
income_mapping = {
    "$30K - $50K": 40000,
    "$70K - $90K": 80000,
    "$50K - $70K": 60000,
    "$10K - $30K": 20000,
    "$90K - $110K": 100000,
    "$110K - $130K": 120000,
    "$130K - $150K": 140000,
    "$150K - $170K": 160000,
    "$190K - $210K": 200000,
    "$250K+": 250000
}
data['Average_Income'] = data['metines_pajamos'].map(income_mapping)

# Setting up the plots
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Plot 1: Distribution of Annual Incomes
sns.histplot(data['Average_Income'], ax=axes[0, 0], bins=30, kde=True)
axes[0, 0].set_title('Metinių pajamų paskirstymas')
axes[0, 0].set_xlabel('Vidutinės pajamos')
axes[0, 0].set_ylabel('Kiekis')

# Plot 2: Gender Distribution
sns.countplot(data=data, x='lytis', ax=axes[0, 1])
axes[0, 1].set_title('Pasiskirstymas pagal lytį')
axes[0, 1].set_xlabel('Lytis')
axes[0, 1].set_ylabel('Kiekis')

# Plot 3: Number of Children per Family
sns.countplot(data=data, x='vaiku_skaicius', ax=axes[1, 0])
axes[1, 0].set_title('Vaikų skaičius šeimose')
axes[1, 0].set_xlabel('Vaikų skaičius')
axes[1, 0].set_ylabel('Kiekis')

# Plot 4: Clients by Country
sns.countplot(data=data, y='kliento_salis', ax=axes[1, 1])
axes[1, 1].set_title('Klientai pagal šalis')
axes[1, 1].set_xlabel('Kiekis')
axes[1, 1].set_ylabel('Šalis')

# Plot 5: Marital Status
sns.countplot(data=data, x='seimynine_padetis', ax=axes[2, 0])
axes[2, 0].set_title('Šeimyninė padėtis')
axes[2, 0].set_xlabel('Šeimyninė padėtis')
axes[2, 0].set_ylabel('kiekis')

# Plot 6: Number of Children in Families by Gender
sns.boxplot(x='lytis', y='vaiku_skaicius', data=data, ax=axes[2, 1])
axes[2, 1].set_title('Vaikų skaičius šeimose pagal lytį')
axes[2, 1].set_xlabel('Lytis')
axes[2, 1].set_ylabel('Vaikų skaičius')

# Adjusting layout
plt.tight_layout()
plt.show()
```
