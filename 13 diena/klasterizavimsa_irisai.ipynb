```python
# Klasterizavimas k-mean. irisai
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/moksliukas/VMI_data/main/13%20diena/data/iris.csv')
y=df
```

```python
df=df.drop(columns='Name')
df
```

```python
plt.scatter(df['SepalLength'],df['SepalWidth'])
```

```python
sse=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
    
plt.plot(range(1,11), sse)
plt.show
```

```python
sil=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(df)
    sil.append(silhouette_score(df, kmeans.labels_))
    
plt.plot(range(2,11), sil)
plt.show
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
y=kmeans.predict(df)
```

```python
plt.scatter(df['SepalLength'],df['SepalWidth'], c=y, cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, alpha=0.5)
plt.show
```
