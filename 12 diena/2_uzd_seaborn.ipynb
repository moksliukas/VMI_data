```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ukrauti Titanic duomen rinkin
titanic = sns.load_dataset("titanic")

# 1. Boxplot: Keleivi Amiaus Skirstymas pagal Bilieto Klas
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='class', y='age')
plt.title("Keleivi Amiaus Skirstymas pagal Bilieto Klas")
plt.show()

```

```python

# 2. Scatterplot: Amius prie Kain
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='class')
plt.title("Amius prie Kain")
plt.show()

```

```python
# 3. Pairplot: Santykiai tarp Keleivi Amiaus, Klass ir Ilikimo
sns.pairplot(titanic, vars=['age', 'fare'], hue='survived')
plt.show()
```

```python

# 4. Heatmap: Koreliacija tarp Skirting Savybi

correlation = titanic.corr(numeric_only=True)
sns.heatmap(correlation, annot=True)
plt.title("Koreliacija tarp Skirting Savybi")
plt.show()
```
