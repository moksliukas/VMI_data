```python
import seaborn as sns
import matplotlib.pyplot as plt

# Užkrauti Titanic duomenų rinkinį
titanic = sns.load_dataset("titanic")

# 1. Boxplot: Keleivių Amžiaus Skirstymas pagal Bilieto Klasę
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='class', y='age')
plt.title("Keleivių Amžiaus Skirstymas pagal Bilieto Klasę")
plt.show()

```

```python

# 2. Scatterplot: Amžius prieš Kainą
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='class')
plt.title("Amžius prieš Kainą")
plt.show()

```

```python
# 3. Pairplot: Santykiai tarp Keleivių Amžiaus, Klasės ir Išlikimo
sns.pairplot(titanic, vars=['age', 'fare'], hue='survived')
plt.show()
```

```python

# 4. Heatmap: Koreliacija tarp Skirtingų Savybių

correlation = titanic.corr(numeric_only=True)
sns.heatmap(correlation, annot=True)
plt.title("Koreliacija tarp Skirtingų Savybių")
plt.show()
```
