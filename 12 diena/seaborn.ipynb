```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# loading dataset 
data = sns.load_dataset("iris") 

def plot(): 
	sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# setting the default color palette 
sns.set_palette('autumn') 
plt.subplot(211) 

# plotting with the color palette 
# as vlag 
plot() 

# setting another default color palette 
sns.set_palette('Accent') 
plt.subplot(212) 
plot() 

plt.show()
```

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 


# loading dataset 
data = sns.load_dataset("iris") 

def graph(): 
	sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# Creating a new figure with width = 5 inches 
# and height = 4 inches 
fig = plt.figure(figsize =(9, 7)) 

# Creating first axes for the figure 
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# plotting the graph 
graph() 

# Creating second axes for the figure 
ax2 = fig.add_axes([0.6, 0.5, 0.3, 0.3]) 

# plotting the graph 
graph() 

plt.show()
```

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# loading dataset 
data = sns.load_dataset("iris") 

def graph(ax): 
    sns.lineplot(x="sepal_length", y="sepal_width", data=data, ax=ax) 

# adding the subplots 
axes1 = plt.subplot2grid((7, 1), (0, 0), rowspan=1)
graph(axes1)

axes2 = plt.subplot2grid((7, 1), (3, 0), rowspan=1)
graph(axes2)

plt.tight_layout()  # Pridkite i eilut, kad pagerintumte idstym
plt.show()
```
