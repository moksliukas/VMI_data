```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm 
df = pd.read_csv('data/Advertising.csv', index_col=0) 
X = df[['TV', 'Radio','Newspaper']] 
y = df['Sales'] 
df.head()
```

```python
df.corr()
```

```python
X = sm.add_constant(X) 
est = sm.OLS(y, X).fit() 
est.summary()
```
