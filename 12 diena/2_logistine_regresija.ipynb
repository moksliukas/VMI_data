```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rashida048/Datasets/master/Heart.csv')
df
```

```python
df=df.drop(columns='Unnamed: 0')
df
```

```python
df['ChestPain']=df['ChestPain'].astype('category')
df['ChestPain']=df['ChestPain'].cat.codes
df
```

```python
df['Thal']=df['Thal'].astype('category')
df['Thal']=df['Thal'].cat.codes

df['AHD']=df['AHD'].astype('category')
df['AHD']=df['AHD'].cat.codes
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
X=df.drop(columns='AHD')
y=df['AHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Normalizacija. sklearn.preprocessing.StandarScaler (z=(x-u)/s)
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()

X_train_scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled=scaler.fit_transform(X_test)
```

```python
print(X_train_scaled.shape, y_train.shape)  # Pvz., (160, 2) ir (160,)
print(X_test_scaled.shape, y_test.shape)   # Pvz., (40, 2) ir (40,)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()

model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score

predictions=model.predict(X_test_scaled)
y_train.shape
```

```python
# Tikslumo įvertinimas
accuracy = accuracy_score(y_test, predictions)
print(f'Modelio tikslumas: {accuracy:.2f}')



# log_reg.score(X_train_scaled, y_train)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
conf_matrix = confusion_matrix(y_test, predictions)

# Apskaičiuojami vertinimai
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
f1 = f1_score(y_test, predictions)

# Grafiškai pavaizduojama teisingumo matrica
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Prognozuotos Reikšmės')
plt.ylabel('Tikrosios Reikšmės')
plt.title('Klaidų Matrica')
plt.show()
```

```python
# Atspausdinami vertinimai
print(f'Modelio tikslumas (Accuracy): {accuracy:.2f}')
print(f'Modelio precizija (Precision): {precision:.2f}')
print(f'Modelio atitikimas (Recall): {recall:.2f}')
print(f'Modelio F1 įvertis: {f1:.2f}')
```
