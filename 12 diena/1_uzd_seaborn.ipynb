```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ukrauti Titanic duomen rinkin
titanic = sns.load_dataset("titanic")

# 1. Amiaus Paskirstymas pagal Lyt
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x="age", hue="sex", multiple="stack")
plt.title("Amiaus Paskirstymas pagal Lyt")
plt.show()

# 2. Ilikimo Lygis pagal Bilieto Klas
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic, x="class", y="survived")
plt.title("Ilikimo Lygis pagal Bilieto Klas")
plt.ylabel("Ilikimo Tikimyb")
plt.show()

# 3. Keleivi Skaiius pagal Ivykimo Uost
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x="embark_town")
plt.title("Keleivi Skaiius pagal Ivykimo Uost")
plt.show()
```
