```python
import seaborn as sns
import matplotlib.pyplot as plt

# Užkrauti Titanic duomenų rinkinį
titanic = sns.load_dataset("titanic")

# 1. Amžiaus Paskirstymas pagal Lytį
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x="age", hue="sex", multiple="stack")
plt.title("Amžiaus Paskirstymas pagal Lytį")
plt.show()

# 2. Išlikimo Lygis pagal Bilieto Klasę
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic, x="class", y="survived")
plt.title("Išlikimo Lygis pagal Bilieto Klasę")
plt.ylabel("Išlikimo Tikimybė")
plt.show()

# 3. Keleivių Skaičius pagal Išvykimo Uostą
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x="embark_town")
plt.title("Keleivių Skaičius pagal Išvykimo Uostą")
plt.show()
```
