```python
# Bibliotekų iškvietimas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Formuojame duomenų rinkinį
medziai=[[35, 8], [49,9], [27,7], [33,6], [60,13], [21,7], [45,11], [51,12]]
df = pd.DataFrame(medziai, columns=['Medžio aukštis', 'Kamieno diametras'])
```

```python
# Braižome diagramą
plt.scatter(df['Medžio aukštis'], df['Kamieno diametras'])
plt.title('Medžio Aukštis vs Kamieno Diametras')
plt.xlabel('Medžio Aukštis (cm)')
plt.ylabel('Kamieno Diametras (cm)')
plt.show()
```

```python
# Paskaičiuojame koreliacija
df.corr(method='pearson')
```
